```python
num_vars = int(input("변수의 개수를 입력하세요 (예: 3, 4, 5, 6): "))
minterms_input = input(f"minterms를 쉼표로 구분하여 입력하세요 (예: 1,3,7): ")
minterms = [int(minterm.strip()) for minterm in minterms_input.split(',')]

dont_cares_input = input(f"don't care term을 쉼표로 구분하여 입력하세요 (예: 5,6,14): ")
dont_cares = [int(dont_care.strip()) for dont_care in dont_cares_input.split(',')] if dont_cares_input else []

all_terms = minterms + dont_cares

def get_binary_representation(num, num_vars):
    return format(num, f'0{num_vars}b')

def group_minterms_by_ones(terms, num_vars):
    groups = {}
    for term in terms:
        binary = get_binary_representation(term, num_vars)
        num_ones = binary.count('1')
        if num_ones not in groups:
            groups[num_ones] = []
        groups[num_ones].append(binary)
    return groups

def find_combinable_pairs(groups):
    new_groups = {}
    checked_pairs = set()
    combined_terms = set()

    keys = sorted(groups.keys())
    for i in range(len(keys) - 1):
        group_a = groups[keys[i]]
        group_b = groups[keys[i + 1]]

        for term_a in group_a:
            for term_b in group_b:
                diff_count = sum(1 for a, b in zip(term_a, term_b) if a != b)
                if diff_count == 1:
                    combined_term = ''.join('_' if a != b else a for a, b in zip(term_a, term_b))
                    if combined_term not in new_groups:
                        new_groups[combined_term] = []
                    new_groups[combined_term].append((term_a, term_b))
                    combined_terms.add(term_a)
                    combined_terms.add(term_b)

    return new_groups, combined_terms

def quine_mccluskey(minterms, dont_cares, num_vars):
    current_groups = group_minterms_by_ones(all_terms, num_vars)
    all_combinations = []
    prime_implicants = []

    while current_groups:
        print("\n현재 그룹핑 상태:")
        for key, value in current_groups.items():
            print(f"{key} 비트의 1을 가진 항들: {value}")

        new_groups, combined_terms = find_combinable_pairs(current_groups)
        all_combinations.append(current_groups)

        # 결합되지 않은 항들만 추출하여 Prime Implicant로 유지
        for group in current_groups.values():
            for term in group:
                if term not in combined_terms and term not in prime_implicants:
                    prime_implicants.append(term)

        # 새 그룹으로 갱신
        current_groups = {}
        for key, value in new_groups.items():
            num_ones = key.count('1')
            if num_ones not in current_groups:
                current_groups[num_ones] = []
            current_groups[num_ones].append(key)

    # 모든 단계의 그룹핑 결과 출력
    print("\n모든 단계의 그룹핑 결과:")
    for idx, combination in enumerate(all_combinations):
        print(f"\n단계 {idx + 1}:")
        for key, value in combination.items():
            print(f"{key} 비트의 1을 가진 항들: {value}")

    # 최종 Prime Implicant 출력
    print("\n최종 Prime Implicant:")
    for prime in prime_implicants:
        print(prime)

    # EPI와 PI 구분 및 출력
    essential_prime_implicants = []
    covered_minterms = set()
    for minterm in minterms:
        count = 0
        epi_candidate = None
        for prime in prime_implicants:
            if is_covered(minterm, prime):
                count += 1
                epi_candidate = prime
        if count == 1 and epi_candidate not in essential_prime_implicants:
            essential_prime_implicants.append(epi_candidate)
            covered_minterms.update(get_covered_minterms(epi_candidate, minterms))

    print("\nEssential Prime Implicants (EPI):")
    for epi in essential_prime_implicants:
        print(epi)

    print("\nNon-Essential Prime Implicants (PI):")
    non_essential_prime_implicants = [pi for pi in prime_implicants if pi not in essential_prime_implicants]
    for pi in non_essential_prime_implicants:
        print(pi)

    # Prime Implicant Chart 생성 및 사용하여 최소한의 SOP 표현식 도출
    minimal_sop_expressions = find_all_minimal_sop_with_chart(essential_prime_implicants, non_essential_prime_implicants, minterms)
    print("\n모든 가능한 Minimal SOP 표현식:")
    for expression in minimal_sop_expressions:
        print(expression)

def is_covered(minterm, implicant):
    return all(m == i or i == '_' for m, i in zip(get_binary_representation(minterm, len(implicant)), implicant))

def get_covered_minterms(implicant, minterms):
    covered = []
    for minterm in minterms:
        if is_covered(minterm, implicant):
            covered.append(minterm)
    return covered

def generate_prime_implicant_chart(prime_implicants, minterms):
    chart = {}
    for minterm in minterms:
        chart[minterm] = []
        for prime in prime_implicants:
            if is_covered(minterm, prime):
                chart[minterm].append(prime)
    return chart

def find_all_minimal_sop_with_chart(epi_list, pi_list, minterms):
    # Step 1: Create Prime Implicant Chart
    prime_implicants = epi_list + pi_list
    chart = generate_prime_implicant_chart(prime_implicants, minterms)

    # Step 2: Add all EPIs to the solution
    initial_solution = list(epi_list)
    remaining_minterms = set(minterms) - set(get_covered_minterms_list(epi_list, minterms))

    # Step 3: Use all combinations of PIs to cover the remaining minterms
    from itertools import combinations
    all_combinations = []
    if not remaining_minterms:
        all_combinations.append((initial_solution, generate_sop_expression(initial_solution, [], minterms)))
    else:
        for r in range(1, len(pi_list) + 1):
            for combination in combinations(pi_list, r):
                solution = initial_solution + list(combination)
                covered_minterms = set(get_covered_minterms_list(solution, minterms))
                if remaining_minterms.issubset(covered_minterms):
                    all_combinations.append((solution, generate_sop_expression(solution, [], minterms)))

    # Find the minimum number of terms in all possible solutions
    if all_combinations:
        min_terms = min(len(combination[0]) for combination in all_combinations)
        minimal_expressions = [expression for solution, expression in all_combinations if len(solution) == min_terms]
    else:
        minimal_expressions = []

    # Return all possible minimal SOP expressions with the least number of terms
    return minimal_expressions

def get_covered_minterms_list(implicants, minterms):
    covered = []
    for implicant in implicants:
        covered.extend(get_covered_minterms(implicant, minterms))
    return covered

def generate_sop_expression(implicants, unused, minterms):
    terms = []
    for implicant in implicants:
        term = ""
        for idx, bit in enumerate(implicant):
            variable = chr(97 + idx)  # 'a', 'b', 'c', ...
            if bit == '0':
                term += f"{variable}'"
            elif bit == '1':
                term += f"{variable}"
        terms.append(term)
    return ' + '.join(terms)

# Quine-McCluskey 알고리즘 실행
quine_mccluskey(minterms, dont_cares, num_vars)
```
